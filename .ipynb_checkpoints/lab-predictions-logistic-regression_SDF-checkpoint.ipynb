```python
#Create a query to get the total amount of rentals in June for each film.
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

query = '''select f.film_id,f.title, count(f.film_id) as june_counts
from film f
left join inventory i
on f.film_id = i.film_id
left join rental r
on i.inventory_id = r.inventory_id
where rental_date like '2005-06%%'
group by f.film_id, f.title;'''

june_count = pd.read_sql_query(query, engine)
june_count.head(5)


```

```python
#Do the same with July.
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

query = '''select f.film_id,f.title, count(f.film_id) as july_counts
from film f
left join inventory i
on f.film_id = i.film_id
left join rental r
on i.inventory_id = r.inventory_id
where rental_date like '2005-07%%'
group by f.film_id, f.title;'''

july_count = pd.read_sql_query(query, engine)
july_count.head(5)

```

```python
#Create a new column containing (Yes/No) for each film whether or not the number of monthly rentals in July was bigger than in June. 
#Your objective will be to predict this new column.
```

```python
#if we want details only on the returned orders, we do joint
june_july_count = pd.merge(left = june_count, # LEFT TABLE
                                 right = july_count, # RIGHT TABLE
                                 how = 'inner', # TYPE OF JOIN
                                 left_on = "title", # COMMON COLUMN IN LEFT TABLE
                                 right_on= "title") # COMMON COLUMN IN RIGTH TABLE

june_july_count.head()
```

```python
june_july_count=june_july_count.drop("film_id_y", axis=1)
```

```python
june_july_count['july_vs_june']= june_july_count['july_counts']>june_july_count['june_counts']
june_july_count
june_july_count[['july_vs_june']].value_counts()
#predict july_vs_june
```

```python
y = june_july_count['july_vs_june']
X = june_july_count.drop(['july_vs_june'], axis=1)
```

```python
display(X.head())
display(y.head())
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = np.object)
display(X_num.head(4))
display(X_cat.head(4))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
#encode some categorical variables 
encoder = OneHotEncoder(drop='first')
encoder.fit(X_cat)
X_cat_encoded = encoder.transform(X_cat).toarray()
cols = encoder.get_feature_names(input_features=X_cat.columns)
X_cat_encoded_df = pd.DataFrame(X_cat_encoded, columns=cols)

X_cat_encoded_df.head()
```

```python
X_num_cat = pd.concat([ X_num,X_cat_encoded_df], axis=1)
X_num_cat.head()
#everything is number
```

```python
sns.displot(X_num_cat['film_id_x'])
plt.show()
```

```python
transformer = MinMaxScaler().fit(X_num_cat)
X_num_cat_norm = transformer.transform(X_num_cat)
display(X_num_cat_norm)
print(X_num_cat_norm.shape)
X_num_cat_norm_df = pd.DataFrame(X_num_cat_norm, columns=X_num_cat.columns)
X_num_cat_norm_df
```

```python
sns.displot(X_num_cat_norm_df['film_id_x'])
plt.show()
```

```python
#Create a logistic regression model to predict this new column from the cleaned data.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_num_cat_norm_df, y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
#How you create a logistic regrassion model
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='ovr')
#solver= 

classification.fit(X_train, y_train)
```

```python
y_train_pred = classification.predict(X_train)
print("The first predictions on the TRAIN set are: ",y_train_pred[:10])
classification.score(X_train, y_train)
```

```python
y_test_pred = classification.predict(X_test)
print("The first predictions on the TEST set are: ",y_test_pred[:10])
classification.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.metrics import classification_report, f1_score, cohen_kappa_score
```

```python
# Precission score. = TP/(TP + FP)
#If we're using labels instead of numbers for the classes to predict, we need to
# provide what class is the "positive" and which is the "negative"
print("The precission in the TRAIN set is: {:.3f}".format(precision_score(y_train, y_train_pred, pos_label="yes")))
print("The precission in the TEST  set is: {:.3f}".format(precision_score(y_test, y_test_pred, pos_label="yes")))
```

```python
# Recall score = TP / ( TP + FN)
print("The recall in the TRAIN set is: {:.3f}".format(recall_score(y_train, y_train_pred, pos_label="yes")))
print("The recall in the TEST  set is: {:.3f}".format(recall_score(y_test,  y_test_pred, pos_label="yes")))
```
